# Environmental setup

1. Install Python by downloading latest version (3.10.6) from python.org
2. In terminal, using pip, install MatPlotLib and JupyterNotebook

```python
> pip install matplotlib
> pip install jupyter
```

3. Run terminal, go to python workspace folder
4. Run python

```python
> python
```

5. Run sample code at the prompt

```python
>>> from matplotlib import pyplot as plt
subjects = ["English", "Nepali", "Maths"]
marks = [60, 70, 80]
plt.bar(range(len(subjects)), marks)
plt.title("Exam marks")
plt.ylabel("Marks in subjects")
plt.xticks(range(len(subjects)), subjects)
plt.show()
```

6. Exit from the python prompt

```python
>>> exit()
```

7. Run jupyter notebook

```python
> jupyter-notebook
```
