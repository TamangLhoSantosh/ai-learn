## A sample program to visualize data in bar chart

```python
# Import pyplot function from matplotlib library, then use 'plt' as an alias 
from matplotlib import pyplot as plt

# Using subject names for x-axis. 
subjects = ["English", "Nepali", "Maths"]
# Assigning the subject names for x-axis ticks
plt.xticks(range(len(subjects)), subjects)
# Labeling the x-axis
plt.xlabel("Subjects" + u"\u2192")

# Using marks as numeric data, therefore no need to assign labels for y-axis ticks
marks = [60, 70, 80]
# Labeling the y-axis
plt.ylabel("Marks in subjects" + u"\u2192")

# Plot the bar charg
plt.bar(range(len(subjects)), marks)

# Set title of the bar chart
plt.title("Exam marks")

plt.show()
```

## Reading data from CSV file then plotting a line graph

```python
import pandas as pd
from matplotlib import pyplot as plt

# read csv data into a data frame

df = pd.read_csv('GDP-per-capita-Nepal.csv')
print(df)

# extract value of year with index
years = df.loc[:,"Year"]
print(years)

# extract value of gdp amount with index
gdp = df.loc[:,"Amount"]
print( gdp)

# plot a line graph
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Nepal's per capita GDP over years. Source: World Bank (2022)")
plt.xlabel("Year")
plt.ylabel("GDP (per capita)")

plt.show()
```

## Bar charts

### Grouped bar charts

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
# x-axis data
X = ['Group A','Group B','Group C','Group D']

# y-axis has two set of data
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]

# generating positions for x-axis ticks 
# np.arange() generates evenly spaced values. Same as range(len(X)) but 
# range generates integer where as np.arange generates float value
X_axis = np.arange(len(X))
plt.xticks(X_axis, X)
  
# calculating offsets of the verticle bars then plotting them separately
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')

plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
# it's useful to display legend in the graph
plt.legend()

plt.show()
```

### Stacked bar charts

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the x locations for the groups

# width of the bars
width = 0.35

plt.bar(ind, menMeans, width, color='r')
plt.bar(ind, womenMeans, width,bottom=menMeans, color='b')

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend(labels=['Men', 'Women'])

plt.show()
```

```python
from bokeh.plotting import figure, output_notebook, show

# x-y coordinates for sample data
x = [1.5, 2.5, 1,3]
y = [1, 1.8, 2.5,3.8]

# redirect output to the JupyterNotebook
output_notebook()


# figure with no toolbar and axis ranges of [0,4]
p = figure(
	title='Sample scatter plot',
	plot_height=500, plot_width=600,
	x_axis_label='X Axis', y_axis_label='Y Axis',
	x_range =(0,4), y_range = (0,4),
    toolbar_location=None
)

# Render Glyph
p.circle(x,y,color='Red', size=10)

# Show plot
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

# x-y coordinates for sample data
x = [1.5, 2.5, 1,3]
y = [1, 1.8, 2.5,3.8]

# Redirect output to a file
output_file('sample-scatter-plot.html', title='Circle Glyphs')

# figure with no toolbar and axis ranges of [0,4]
p = figure(
	title='Sample scatter plot',
	plot_height=500, plot_width=600,
	x_axis_label='X Axis', y_axis_label='Y Axis',
	x_range =(0,4), y_range = (0,4),
    toolbar_location=None
)

# Render Glyph
p.circle(x,y,color='Red', size=10)

# Show plot
show(p)
```
